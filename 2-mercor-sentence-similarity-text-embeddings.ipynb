```python
import os
import warnings
import numpy as np
import pandas as pd

from tqdm import tqdm
warnings.filterwarnings('ignore')
from sklearn.metrics.pairwise import cosine_similarity
```

```python
mercor = pd.read_csv('/kaggle/input/mercor-clothing/mercor_clothing_data.csv')
mercor.head()
```

```python
!pip install sentence-transformers >> /dev/null

from sentence_transformers import SentenceTransformer

model = SentenceTransformer('bert-base-nli-mean-tokens')
model
```

```python
sentences = mercor['description'].values.tolist()
sentence_embeddings = model.encode(sentences)
```

```python
sentence_embeddings_df = pd.DataFrame(sentence_embeddings)
sentence_embeddings_df.to_csv('mercor_clothing_text_embeddings.csv', index=False)
```

```python
import numpy as np
for i in tqdm(sentence_embeddings_df.columns):
    sentence_embeddings_df[i] = sentence_embeddings_df[i].astype(np.float16)
```

```python
sentence_embeddings_df['category'] = mercor['category']
sentence_embeddings_df['product_url'] = mercor['product_url']
sentence_embeddings_df['gender'] = mercor['gender']
```

```python
sentence_embeddings_df.to_csv('mercor_clothing_text_embedding.csv', index=False)
```
